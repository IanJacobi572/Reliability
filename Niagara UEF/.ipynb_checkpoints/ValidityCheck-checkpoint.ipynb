```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
fileNameC = 'Takashi/Results/103890462-3-1S1-1160.SM22014.csv'
df = pd.read_csv(fileNameC)
```

```python
def find_nearest(array, value):
    idx = np.searchsorted(array, value, side="left")
    try: 
        if isclose(array[idx], value, abs_tol=1):
            return idx
    except:
        pass
    return idx-1
```

```python
df['Validity']='Valid'

#Check if Room Temperature in range 68+/-2.05 and draws are 14 or (more?)
if (df['Ambient TC'].astype('float64') > 70).any() or (df['Ambient TC'] < 65).any() or max(df['Draw no'].astype('int64').values)<14:
    df['Validity']='Not Valid'
#If it's still valid then check Tin during draws if it's in range 56-60
else:
    for i in range(1,df['Draw no'].max()+1):
        t1 = df.loc[df['Draw no']==i, 'TimeStamp (sec)'].iloc[0]
        idx1 = find_nearest(df['TimeStamp (sec)'].values, t1+5)
        idx2 = df.index[df['Draw no']==i][-1]
        x = (df.loc[idx1:idx2, 'Tin']<56).any()
        if (df.loc[idx1:idx2, 'Tin']<56).any() or (df.loc[idx1:idx2, 'Tin']>60).any():
            df['Validity']='Not Valid'
            break
```

```python
df['Validity'].unique()
```

```python
plt.plot(df['TimeStamp (sec)']/3600, df['Ambient TC'].astype('float64'), color='b')
plt.axhline(y=70, linestyle='dashed', color='r')
plt.axhline(y=65, linestyle='dashed', color='r')
plt.xlabel('Time (hrs)')
plt.show()
```

```python
plt.plot(df['Draw no'].loc[df['Draw no']!=0], df['Tin'].loc[df['Draw no']!=0].astype('float64'), color='b')
plt.axhline(y=60, linestyle='dashed', color='r')
plt.axhline(y=56, linestyle='dashed', color='r')
plt.xlabel('Draws')
plt.show()
```
