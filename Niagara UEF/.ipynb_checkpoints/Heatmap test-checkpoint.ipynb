```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('RQ3 ICN 10196 RTGH-S11IS1-0907.SM22014.csv')
```

```python
df.columns
```

```python
df= df[['Ambient TC', 'Gas TC', 'Barometer', 'Gas ascf',
       'WattHrs', 'Water (Gallons)', 'Water Flow (GPM)', 'Tin',
       'Tout', 'Drawn Water (Gallons)', 'M']].copy()
```

```python
fig, ax = plt.subplots(figsize=(14,8))
correlation_matrix = df.corr().round(2)
sns.heatmap(data=correlation_matrix, annot=True)
```
